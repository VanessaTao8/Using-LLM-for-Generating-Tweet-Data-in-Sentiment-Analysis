```python
from openai import OpenAI
import pandas as pd
import re
import os
```

```python
client = OpenAI(
    # defaults to os.environ.get("OPENAI_API_KEY")
    api_key = os.environ.get("OPENAI_API_KEY"),
)
model_name = "gpt-3.5-turbo-1106"

# read csv file
csv_path = 'dreaddit-train.csv' 
data = pd.read_csv(csv_path)
```

```python

user_messages = [{"role":"system", "content":"You are a helpful assistent. Please help the user to generate a dataset for stress analysis in social media. Generate 100 rows of new data each time."}]
message = 'Please generate more data exactly in this format: \n{text:"{text}",stress:"{stress}"}, where {text} is the content of the post in Reddit with at most 5 sentences, {stress} is a binary stress label, 0 means non-stress and 1 means stress.\n The following are some examples.\n'

# integragate original data 
for index, row in data[:50].iterrows():
    m = f'text:"{row["text"]}",stress:"{row["label"]}"\n' 
    message += m

user_messages.append({"role":"user", "content":message})
```

```python
response = client.chat.completions.create(
    model = model_name,
    messages = user_messages,
    temperature = 1,
    max_tokens = 2048,
    top_p = 1,
    frequency_penalty = 0,
    presence_penalty = 0,
    n = 1
)
```

```python
generated_content = response.choices[0].message.content
print(generated_content)
```
